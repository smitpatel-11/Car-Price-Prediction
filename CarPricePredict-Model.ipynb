```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Cardetails.csv")
df.head(5)
```

```python
df.drop(columns=["torque"],inplace=True)
```

```python
# 23.4 kmpl	  1248 CC	74 bhp
# Numerical conversion of this string values

df['mileage'] = df['mileage'].astype(str).str.replace(' kmpl', '', regex=False)
df['mileage'] = df['mileage'].str.replace(' km/kg', '', regex=False)
df['mileage'] = pd.to_numeric(df['mileage'], errors='coerce')

df['engine'] = df['engine'].astype(str).str.replace(' CC', '', regex=False)
df['engine'] = pd.to_numeric(df['engine'], errors='coerce')

df['max_power'] = df['max_power'].astype(str).str.replace(' bhp', '', regex=False)
df['max_power'] = pd.to_numeric(df['max_power'], errors='coerce')
```

```python
# owner
# First Owner             5289
# Second Owner            2105
# Third Owner              555
# Fourth & Above Owner     174
# Test Drive Car             5

own={
    'Test Drive Car':0,
    'First Owner':1,
    'Second Owner':2,
    'Third Owner':3,
    'Fourth & Above Owner':4
}
df["owner"]=df["owner"].map(own)
```

```python
# seller_type
# Individual          6766
# Dealer              1126
# Trustmark Dealer     236

seller={
    'Individual':0,
    'Dealer':1,
    'Trustmark Dealer':2
}

df["seller_type"]=df["seller_type"].map(seller)
```

```python
# transmission
# Manual       7078
# Automatic    1050

df['transmission'] = df['transmission'].map({'Manual': 0, 'Automatic': 1})
```

```python
df.groupby('fuel')['selling_price'].mean().sort_values(ascending=False)
```

```python
fuel={
    "LPG":0,
    "CNG":1,
    "Petrol":2,
    "Diesel":3
}

df["fuel"]=df["fuel"].map(fuel)
```

```python
df["Company"]=df["name"].str.split(" ").str[0]
df['model']=df["name"].str.split(" ").str.slice(2,).str.join(" ")
df["name"]=df["Company"]=df["name"].str.split(" ").str[1]
df.sample(10)
```

```python
df["year"]=2024-df["year"]
df.dropna(inplace=True)
df.sample(10)
```

```python
df=df.drop(columns=['model'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns=["selling_price"]),df["selling_price"],test_size=0.2,random_state=42)

preprocessor = ColumnTransformer(
    transformers=[
        ('carVals', OneHotEncoder(handle_unknown='ignore'), ["name","Company"]),
        ('StandardScaler',StandardScaler(),["year","km_driven","mileage","engine","max_power","seats"])
        
    ],
    remainder='passthrough'
)
```

```python
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.model_selection import cross_val_score
import optuna
```

```python
models={
    'LinearRegression':LinearRegression(),
    'RandomForestRegressor':RandomForestRegressor(),
    'GradientBoostingRegressor':GradientBoostingRegressor(),
    'XGBRegressor':XGBRegressor()
}

for name, model in models.items():
    pipe = Pipeline([
        ('preprocessor', preprocessor),
        ('model', model)
    ])
    
    scores=cross_val_score(pipe,x_train,y_train,cv=5,scoring='r2')
    print(f"{name} CV mean is {scores.mean():.4f}")
```

```python
def objective(trial):
    n_estimators=trial.suggest_int("n_estimators",100,1000,step=50)
    max_depth=trial.suggest_int("max_depth",3,10)
    learning_rate=trial.suggest_float("learning_rate",0.01,0.3,log=True)

    model=Pipeline([
        ('preprocessor',preprocessor),
        ('regressor',XGBRegressor(
            n_estimators=n_estimators,
            max_depth=max_depth,
            learning_rate=learning_rate,
            objective="reg:squarederror",
            n_jobs=-1
        ))
    ])

    score=cross_val_score(model,x_train,y_train,cv=5,scoring='r2')
    return score.mean()

study=optuna.create_study(direction='maximize')
study.optimize(objective,n_trials=50)
```

```python
print("Best parameters : ",study.best_params)
print("Best CV score : ",study.best_value)
```

```python
best_params=study.best_params

final_model=Pipeline([
    ('preprocessor',preprocessor),
    ('regressor',XGBRegressor(
            n_estimators=best_params['n_estimators'],
            max_depth=best_params['max_depth'],
            learning_rate=best_params['learning_rate'],
            objective="reg:squarederror"
        ))
])

final_model.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score,mean_squared_error

y_pred=final_model.predict(x_test)

print("R2 Score = ",r2_score(y_test,y_pred))
print("RMSE = ",np.sqrt(mean_squared_error(y_test,y_pred)))
```
